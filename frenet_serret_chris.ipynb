```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import frenet_serretII as frenet_serret

from mpl_toolkits.mplot3d import Axes3D

#import importlib
#importlib.reload(frenet_serret)
```

```python
# make some points and tangents analytically:

points = [np.array([np.cos(t), np.sin(t), 0.1*np.sin(5*t)]) for t in np.linspace(0, 2 * np.pi, 30, endpoint=False)]
tangents = [np.array([-np.sin(t), np.cos(t), 0.5*np.cos(5*t)]) for t in np.linspace(0, 2 * np.pi, 30, endpoint=False)]
frames = frenet_serret.computeFrenetSerretFrames(points, tangents)
print(frames)

```

```python
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


xpoints = [frame.position[0] for frame in frames]
ypoints = [frame.position[1] for frame in frames]
zpoints = [frame.position[2] for frame in frames]

ax.plot(xpoints, ypoints, zpoints, marker = 'o')


for frame in frames[0::5]:
    ax.quiver(frame.position[0], 
              frame.position[1], 
              frame.position[2],
              frame.tangent[0],
              frame.tangent[1],
              frame.tangent[2], color='r')
    ax.quiver(frame.position[0], 
              frame.position[1], 
              frame.position[2],
              frame.normal[0],
              frame.normal[1],
              frame.normal[2], color='g')
    ax.quiver(frame.position[0], 
              frame.position[1], 
              frame.position[2],
              frame.binormal[0],
              frame.binormal[1],
              frame.binormal[2], color='k')


```
